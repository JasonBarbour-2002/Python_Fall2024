## Basic Plots

Let's first create some synthetic data to plot. We will use the `numpy` library to generate the data.

````python
import numpy as np 

x = np.arange(-2 * np.pi, 2 * np.pi,0.1)
y = np.cos(x)
````

Line plots are the most common type of plot.

````python
plt.plot(x,y)
plt.show()
````

You can also have a scatter plot:

````python
plt.scatter(x,y)
plt.show()
````

or 

````python
plt.plot(x,y,'o')
plt.show()
````

## Customizing Plots

`matplotlib` allows you to customize your plots in many ways. You can change the color, line style, marker style, and many other properties. You can also add labels, titles, and legends to your plots. Let's see some examples:

````python
x = np.arange(-3,3.2,0.2)
y = x**2
plt.plot(x, y, color='red', linestyle=':', marker='*', mfc="black", mec="black", label='y = x^2')
plt.title('Graph of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
````

You can acheive the same by using the `fmt` (format) "[marker][line][color]" argument in the `plot` function:

````python
plt.plot(x, y, "*:r", mfc="k", mec="k",label='y = x^2')
plt.title('Graph of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
````

You can also change the x and y limits of the plot to zoom in or out of the data:

````python
plt.plot(x,y, "*:r", mfc="k", mec="k",label='y = x^2')
plt.title('Graph of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.ylim(0,10)
plt.xlim(0,3)
plt.show()
````

You can also plot multiple plots on the same figure

````python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.5)
y = np.cos(x)
z = np.sin(x)
w = x ** 3 / 40
````

````python
plt.plot(x,y,'*:r',label='y = cos(x)', mfc='r', mec='r')
plt.plot(x,z,'o--b',label='z = sin(x)', mfc='b', mec='b')
plt.plot(x,w,'s-.g',label='w = x^3', mfc='g', mec='g')
plt.title('Graph of y = cos(x), z = sin(x), w = x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
````

To save a plot to a file, you can use the `savefig` function:

````python
plt.plot(x,y,'*:r',label='y = cos(x)', mfc='r', mec='r')
plt.plot(x,z,'o--b',label='z = sin(x)', mfc='b', mec='b')
plt.plot(x,w,'s-.g',label='w = x^3', mfc='g', mec='g')
plt.title('Graph of y = cos(x), z = sin(x), w = x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.savefig('graph.png')
plt.show()
````

### Subplots

This is usefull to plot multiple plots in different subplots

````python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.cos(x)
z = np.sin(x)
w = x ** 3 / 40
k = np.exp(x) / 100
fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,0].plot(x,y, 'r', label='y = cos(x)')
ax[0,0].set_title('y = cos(x)')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')
ax[0,0].grid()
ax[0,0].legend(loc='upper right')

ax[0,1].plot(x,z, 'b', label='z = sin(x)')
ax[0,1].set_title('z = sin(x)')
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('z')
ax[0,1].grid()
ax[0,1].legend(loc='upper right')

ax[1,0].plot(x,w, 'g', label='w = x^3')
ax[1,0].set_title('w = x^3')
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('w')
ax[1,0].grid()
ax[1,0].legend(loc='upper right')

ax[1,1].plot(x,k, 'y', label='k = exp(x)')
ax[1,1].set_title('k = exp(x)')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('k')
ax[1,1].grid()
ax[1,1].legend(loc='upper right')


plt.show()
````

### Histograms

Histograms are a great way to visualize the distribution of a dataset. You can use the `hist` function to create a histogram.

````python
mu = 0
sigma = 1
data = np.random.normal(mu, sigma, 1000)
x = np.arange(-5,5,0.1)

fig = plt.figure()
plt.hist(data, bins=100, color='red', alpha=0.5, label='data', density=True)
plt.plot(x,1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2)), 'k--', label='normal distribution')
plt.title('Histogram of data')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

### 3D Plots

You can also create 3D plots using the `mplot3d` toolkit.

````python
%matplotlib widget
z = np.arange(-4 * np.pi, 4 * np.pi, 0.3)
y = np.sin(z)
x = np.cos(z)


# 3-d plot
fig, ax = plt.subplots(figsize=(10,10), subplot_kw={'projection':'3d'})
ax.plot(x,y,z, 's--r', label='y = sin(z) and x = cos(z)', mfc='k', mec='k', ms=5, lw=1)
ax.set_title('3D plot of y = sin(z) and x = cos(z)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
plt.show()
````

You can also plot surfaces: 

````python
%matplotlib widget
fig, ax = plt.subplots(figsize=(10,10),subplot_kw={"projection": "3d"})
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.6)
fig.colorbar(surf, shrink=0.5, aspect=20)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
````

### Animations

You can create animations using the `animation` module.

````python
from matplotlib.animation import FuncAnimation
````

Note that in all previous plots the `%matplotlib widget` magic command was not required. However, for animations, you need to use it.

````python
%matplotlib widget
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r')
ln2, = plt.plot([], [], 'b')

def init():
    ln.set_data([], [])
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x + frame)
    z = np.cos(x - frame)
    ln.set_data(x, y)
    ln2.set_data(x, z)
    return ln,ln2,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True, interval=10)
plt.title('Sine and Cosine wave')
plt.show()
````

### Reading and Modifying Images

You can also read and modify images using the `image` module.

````python
from matplotlib.image import imread

im1 = imread('Data/Image1.png')
````

````python
type(im1), im1.shape
````

You can also plot the image using the `imshow` function:

````python
fig, ax = plt.subplots()
ax.imshow(im1)
plt.axis('off')
plt.show()
````

To modify the image you can of course use the full power of `numpy`. For example, you can convert the image to grayscale:

````python
grey = np.mean(im1, axis=2)
````

````python
plt.imshow(grey, cmap='gray')
plt.axis('off')
plt.show()
````
